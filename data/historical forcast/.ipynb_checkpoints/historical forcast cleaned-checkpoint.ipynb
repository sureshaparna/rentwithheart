```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_excel("Moving annual median rent by suburb and town - March quarter 2025.xlsx", sheet_name=None, header=None, skiprows=1)

all_df = pd.DataFrame()
for name, df in data_df.items():
    header1 = df.iloc[0].astype(str)
    header2 = df.iloc[1].astype(str)
    header = header1 + "_" + header2
    
    df = df.iloc[2:].reset_index(drop=True)
    df.columns = ["region", "suburb"] + list(header[2:])
    
    df = df.replace(r'^\s*$', pd.NA, regex=True).replace("-", pd.NA).dropna(axis=1, how='all')
    df["region"] = df["region"].ffill()
    df["type"] = name
    df = df[df["suburb"] != "Group Total"]

    all_df = pd.concat([all_df, df], ignore_index=True)

all_df.head()
```

```python
final_df = all_df.melt(
    id_vars=["region", "suburb", "type"],
    var_name="variable",
    value_name="value"
)

final_df[["timestamp", "metric"]] = final_df["variable"].str.rsplit("_", n=1, expand=True)
final_df = final_df.pivot_table(
    index=["region", "suburb", "type", "timestamp"],
    columns="metric",
    values="value",
    aggfunc="first"
).reset_index()

final_df["timestamp"] = pd.to_datetime(final_df["timestamp"], format="%b %Y")
final_df = final_df.sort_values(["region", "suburb", "type", "timestamp"])

final_df.head(50)
```

```python
final_df.to_csv("historical_forcast_cleaned.csv", index=False)
```
